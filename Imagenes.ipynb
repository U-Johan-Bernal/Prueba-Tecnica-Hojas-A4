```python
# Libraries
import cv2
from reportlab.pdfgen import canvas
from os import remove


#Height and With of A4 paper sheet

base_width=796
base_height=1123


# Function to evaluate image
    
def route (name):
    
    img = cv2.imread("M:/Pictures/"+name+".jpg")
    img_height, img_width = img.shape[0:2]
    
    # know the pixels of the image 
    # print(img_height, img_width)
        
    #position Y of image in the PDF
    position_height=base_height-int(img_height)
    
    # Comparison of pixels referring to A4 paper sheet
    if (img_height<=base_height and img_width<=base_width):
        
        # save image to PDF 
        
        route_pdf="M:/Documents/Johan/Proyectos/"+name+".pdf"
        c = canvas.Canvas("M:/Documents/Johan/Proyectos/"+name+".pdf")
        c.setPageSize((base_width, base_height))
        c.drawImage("M:/Pictures/"+name+".jpg", 0,  int(position_height))
        c.save()
        
        # Recommended pixels for A4 sheet of paper 
        print("The image "+ name + " its recommended pixels are: "+str(int(img_width))+"x"+str(int(img_height)))
        return
        
    elif(img_height>img_width):
        # Modify size height 
        img_width=(img_width*base_height)/img_height
        img_height=base_height 
                
    elif(img_width>img_height):
        # Modify size
        img_height=(img_height*base_width)/img_width
        img_width=base_width
        
    elif(img_width==img_height):
        # Modify size
        img_height=base_width
        img_width=base_width
        
    # Position Y of image in the PDF
    position_height=base_height-int(img_height)
   
    # image creation with new size
    newImg=cv2.resize(img,(int(img_width),int(img_height)))
    
    ##save image to PDF 
    
    cv2.imwrite("M:/Documents/Johan/Proyectos/"+name+".jpg", newImg)
    c = canvas.Canvas("M:/Documents/Johan/Proyectos/"+name+".pdf")
    c.setPageSize((base_width, base_height))
    c.drawImage("M:/Documents/Johan/Proyectos/"+name+".jpg", 0,  int(position_height))
    c.save()
    remove("M:/Documents/Johan/Proyectos/"+name+".jpg")
    
    # Recommended pixels for A4 sheet of paper 
    print("The image "+ name + " its recommended pixels are: "+str(int(img_width))+"x"+str(int(img_height)))
        
# image file name 
route("ejemplo")
route("ejemplo1")
route("ejemplo2")
route("ejemplo3")
route("ejemplo4")
```
